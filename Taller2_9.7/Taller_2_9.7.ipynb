```python
import math
import numpy as np
import matplotlib.pyplot as ptl
%matplotlib inline
    
def grafica2():
    x=np.arange(-2,40,0.1)
    t=(-9.5-0.5*x)/(-1)
    r=(-18.8-1.02*x)/-2
    ptl.plot(x,t,'Blue',label='X2')
    ptl.plot(x,r,'DeepPink',label='X1')
    ptl.title("Ejercicio 9.7", fontsize = 20)
    ptl.legend(loc="upper left")
    ptl.grid(True)
    ptl.show()

grafica2()
```
