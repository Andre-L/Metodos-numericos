```python
import math
import numpy as np
import matplotlib.pyplot as ptl
%matplotlib inline
def grafica():
    x=np.arange(3,6,0.5)
    y=2*x-6
    z=(34-6*x)
    ptl.plot(x,y,'crimson',label='X1')
    ptl.plot(x,z,'SpringGreen',label='X2')
    ptl.title("Ejercicio 9.4", fontsize = 20)
    ptl.legend(loc="upper right")
    ptl.grid(True)
    ptl.show()
grafica()
```
